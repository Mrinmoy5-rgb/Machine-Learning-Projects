```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python

df = pd.read_csv(r"C:\Users\nandi\Mrinmoy_python\Machine-Learning-Projects\INTERNPE_INTERNSHIP\diabetes.csv")

print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:\n", df.head())
```

```python
print("\nStatistical Summary:\n", df.describe())

print("\nClass Distribution:\n", df['Outcome'].value_counts())
```

```python
# Handle Invalid Zero Values
columns_to_clean = ['Glucose', 'BloodPressure',
                    'SkinThickness', 'Insulin', 'BMI']

for col in columns_to_clean:
    df[col] = df[col].replace(0, df[col].median())

print("\nZero values replaced with median.")
```

```python
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print("\nTraining Set Shape:", X_train.shape)
print("Test Set Shape:", X_test.shape)
```

```python
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
print(standardized_data)
```

```python
# Train SVM Model
model = SVC(kernel="linear", class_weight="balanced")
model.fit(X_train, y_train)
```

```python
print(X)
print(y)
```

```python
# Model Evaluation
y_pred = model.predict(X_test)

print("\nTest Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
